```python
import os
from crewai import LLM, Agent, Task, Crew
from dotenv import load_dotenv

# Load variables from .env
load_dotenv()
# Access them using os.getenv
api_key = os.getenv("OPENAI_API_KEY")
```

```python
llm = LLM(model="openai/gpt-5-nano", temperature=1)
llm.call("Who is father of Artificial Intelligence")
```

```python
original_email = "hey team just want to bring all you up to the latest tea in the office is that the manager of Ai department is getting kicked off by the HR ASAP, as sketchy and harrasment work relations with this NOT TO BE DISCLOSED PERSON"
```

```python
email_assistant = Agent(
    role="Email Assistant Agent",
    goal="Improve the Email and make them sound professional and clear",
    backstory="A highgly communication expert skilled in professional email writing.",
    # verbose=True,
    llm=llm,
)

task = Task(
    agent=email_assistant,
    description=f"""take the following email and re write the email in professional and more polished version  {original_email}""",
    expected_output="A professional written email with proper formatting and spelling."
)

crew = Crew(
    agents=[email_assistant],
    tasks=[task],
    # verbose=True,
)

result = crew.kickoff()

print("THIS IS RESULT ======= \n",result)
```
